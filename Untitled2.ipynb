b. Jelaskan bagaimana keberadaan missing values dapat memengaruhi analisis data.
Jawab:

Keberadaan **missing values** dalam dataset dapat memengaruhi analisis data sebagai berikut:

1. **Mengurangi Akurasi Model**: Data yang hilang dapat menurunkan akurasi model karena informasi yang tidak lengkap mempengaruhi analisis hubungan antar variabel.
  
2. **Menimbulkan Bias**: Jika data hilang tidak acak, hal ini bisa menyebabkan bias yang membuat hasil analisis tidak mencerminkan kondisi populasi secara keseluruhan.

3. **Mengurangi Ukuran Sampel**: Menghapus baris atau kolom dengan nilai hilang dapat memperkecil ukuran sampel, yang berdampak pada daya statistik analisis.

4. **Mempengaruhi Statistik Deskriptif**: Missing values dapat mempengaruhi perhitungan statistik seperti rata-rata dan median, sehingga hasilnya bisa tidak tepat.

5. **Merusak Asumsi Model**: Banyak model analisis mengharuskan tidak ada missing values, jadi keberadaan nilai hilang bisa merusak validitas hasil.

Menangani missing values dengan cara yang tepat, seperti imputasi atau penghapusan, sangat penting untuk memastikan hasil analisis tetap akurat dan valid.

Jawab:

Winsorizing (Mengganti Outlier dengan Batas IQR) Karena jika data penting dan tidak boleh dihapus, kita bisa mengganti outlier dengan batas kuartil. Winsorizing digunakan untuk mempertahankan data tanpa menghilangkan outlier, sehingga nilai ekstrem tidak memengaruhi statistik seperti mean dan standar deviasi. Dalam analisis harga produk, metode ini membantu mengoreksi nilai yang tidak logis, seperti harga negatif atau terlalu tinggi, tanpa mengurangi jumlah data atau mengganggu hasil analisis.

#**IQR**


```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # Kuartil 1
    Q3 = data[column].quantile(0.75)  # Kuartil 3
    IQR = Q3 - Q1  # Rentang interkuartil
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Cek jumlah outlier untuk setiap kolom numerik
for col in num_cols:
    outliers = detect_outliers_iqr(data, col)
    print(f"{col}: {len(outliers)} outliers")
```

#**Winsorizing (Mengganti Outlier dengan Batas IQR)**

```python
def winsorize_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])

# Terapkan Winsorizing
for col in num_cols:
    winsorize_iqr(data, col)
```

b. Bandingkan berbagai metode encoding dan pilih metode yang paling sesuai untuk
setap kolom. Jelaskan alasan pemilihan metode tersebut.

Jawab:

One-Hot Encoding unggul karena tidak menganggap adanya urutan dalam kategori, sehingga cocok untuk jumlah kategori yang sedikit. Namun, kelemahannya adalah dapat meningkatkan jumlah fitur secara drastis jika kategori terlalu banyak. Di sisi lain, Ordinal Encoding lebih efisien untuk model yang dapat memahami keterkaitan berurutan antar kategori, tetapi kurang tepat jika kategori tersebut tidak memiliki urutan yang jelas. Sementara itu, Target Encoding lebih sederhana dibandingkan One-Hot Encoding dan dapat menangkap hubungan dengan variabel target, tetapi berisiko menyebabkan overfitting jika tidak diatur dengan baik.

Kolom Kategori: One-Hot Encoding
Karena jumlah kategori tidak terlalu banyak, One-Hot Encoding tetap efisien tanpa meningkatkan dimensionalitas data secara berlebihan.

Negara Asal: One-Hot Encoding
One-Hot Encoding dipilih untuk negara asal karena tidak mengasumsikan urutan antar kategori dan jumlah kategorinya masih wajar, sehingga tidak meningkatkan jumlah fitur secara berlebihan. Metode ini juga menghindari risiko overfitting yang bisa terjadi pada Target Encoding.
